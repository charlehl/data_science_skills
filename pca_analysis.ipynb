# Data Visualization using PCA

```python
from sklearn.preprocessing import StandardScaler
```

```python
x = breast_dataset_df.loc[:, features].values
```

```python
# Normalize the features
x = StandardScaler().fit_transform(x)
```

```python
x.shape
```

```python
# Normalized data should have mean ~ 0, std dev 1
np.mean(x), np.std(x)
```

```python
pd.DataFrame(x).head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca_breast = PCA(n_components=2)
pc_breast = pca_breast.fit_transform(x)
```

```python
pc_breast_df = pd.DataFrame(data=pc_breast, 
                           columns=['pca1', 'pca2'])
```

```python
pc_breast_df.shape
```

```python
print(f'Explained variation per principal component: {pca_breast.explained_variance_ratio_}')
```

```python
info_lost = 1 - pca_breast.explained_variance_ratio_[0] - pca_breast.explained_variance_ratio_[1]
```

```python
print(info_lost)
```

```python
plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component - 1',fontsize=20)
plt.ylabel('Principal Component - 2',fontsize=20)
plt.title("Principal Component Analysis of Breast Cancer Dataset",fontsize=20)
targets = ['Benign', 'Malignant']
colors = ['r', 'g']
for target, color in zip(targets,colors):
    indicesToKeep = breast_dataset_df['label'] == target
    plt.scatter(pc_breast_df.loc[indicesToKeep, 'pca1']
               , pc_breast_df.loc[indicesToKeep, 'pca2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```

# Speeding Up ML Using PCA

```python
x_test
```

```python
# Normalize the testing data for images
x_test = x_test/255.0
```

```python
x_test.shape
```

```python
x_test_flat = x_test.reshape(-1, (32*32*3))
```

```python
x_test_flat.shape
```

```python
# Create PCA model to create components to capture 90% of variance
pca = PCA(0.9)
```

```python
pca.fit(x_train_flat)
```

```python
pca.n_components_
```

```python
train_img_pca = pca.transform(x_train_flat)
test_img_pca = pca.transform(x_test_flat)
```

```python
#import libraries for deep learning model
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from keras.optimizers import RMSprop
```

```python
# convert labels to one-hot encoding
y_train_oh = np_utils.to_categorical(y_train)
y_test_oh = np_utils.to_categorical(y_test)
```

```python
# set params for model
batch_size = 128
num_classes = 10
epochs = 20
```

```python
# Create sequential model
model = Sequential()
model.add(Dense(1024, activation='relu', input_shape=(99,)))
model.add(Dense(1024, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])
```

```python
history = model.fit(train_img_pca, y_train_oh,batch_size=batch_size,epochs=epochs,verbose=1,
                    validation_data=(test_img_pca, y_test_oh))
```

```python
model_nopca = Sequential()
model_nopca.add(Dense(1024, activation='relu', input_shape=(3072,)))
model_nopca.add(Dense(1024, activation='relu'))
model_nopca.add(Dense(512, activation='relu'))
model_nopca.add(Dense(256, activation='relu'))
model_nopca.add(Dense(num_classes, activation='softmax'))
model_nopca.compile(loss='categorical_crossentropy',
              optimizer=RMSprop(),
              metrics=['accuracy'])
```

```python
history_nopca = model_nopca.fit(x_train_flat, y_train_oh,batch_size=batch_size,epochs=epochs,verbose=1,
                    validation_data=(x_test_flat, y_test_oh))
```

```python
score_nopca = model_nopca.evaluate(x_test_flat, y_test_oh)
```

```python
score_nopca
```

```python
# Score on test data for model without PCA
print('Test loss:', score_nopca[0])
print('Test accuracy:', score_nopca[1])
```

```python
score = model.evaluate(test_img_pca, y_test_oh)
```

```python
# Score on test data for model with PCA
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
